```python
#Abstraction
class LibraryItem:
    def __init__(self,item_type,borrow_days):
        self.item_type = item_type
        self.borrow_days = borrow_days
    def calculate_charge(self):
      pass
class Book(LibraryItem):   #Inheritance
    def __init__(self,item_type,borrow_days):
        super().__init__(item_type,borrow_days)
        self._rate_per_day=10
    def calculate_charge(self):
        return self.borrow_days * self._rate_per_day
class magazine(LibraryItem):
    def __init__(self,item_type,borrow_days):
        super().__init__(item_type,borrow_days)
        self._rate_per_day=10
    def calculate_charge(self):
        return self.borrow_days * self._rate_per_day
class LibraryApp:
    def __init__(self):
        self.items = None
    def create_item(self,item_type):#Polymorphism
        if item_type == 'book':
            self.items = Book("Books",5)
        else:
            self.items = magazine("magazine",3)
            return"Item Created"
    def get_charge(self):
      return self.items.calculate_charge()
app = LibraryApp()
print(app.create_item("Book"))
print("Borrowing Charge:", app.get_charge())

app2 = LibraryApp()
print(app2.create_item("Magazine"))
print("Borrowing Charge:", app2.get_charge())



```
