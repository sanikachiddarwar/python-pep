```python
class shape:
  def shape(self,area):
    pass
class circle(shape):
  def shape(self,area):
    print("area of circle",area)
class rectangle(shape):
  def shape(self,area):
    print("area of rectangle",area)
class square(shape):
  def shape(self,area):
    print("area of square",area)
obj=circle()
obj.shape(34)
obj=rectangle()
obj.shape(45)
obj=square()
obj.shape(67)
```
