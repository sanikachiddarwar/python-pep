```python
file=open("student.txt","r")
data=file.read()
print(data)
file.close()
```

```python
file=open("string.txt","w")
content="my name is sanika\n"
file.write(content)
file.write(content.upper())
file.close()
```

```python
file=open("string.txt","r")
data=file.read()
print(len(data))
file.close()
```

```python
file=open("string.txt","r")
data=file.read()
words=data.split()
print(len(words))
print(data.count("is"))
file.close()
```

```python
file=open("string.txt","r")
data=file.read()
count=data.lower().count("a")
print(count)
file.close()
```

```python
file=open("string.txt","r")
data=file.read()
print(data.replace("sanika","harsh"))
file.close()
```

```python
file=open("string.txt","r")
data=file.read()
print(data[::-1])
file.close()
```

```python
file=open("movie.txt","r")
data=file.read()
print(data.replace(" ",""))
file.close()
```

```python
#reversing
file=open("movie.txt","r")
data=file.read()
words=data.split()
for w in words:
  print(w[::-1])
  file.close()
```
