1. generate random number
2. check which number is even/odd
3. round number using math
4. count how many times numbers apper
5. saves the result in the file

```python

import random
import math
import os
import re
from collections import Counter
numbers = [random.randint(1,10) for _ in range(10)]
#even or odd
even_numbers=[]
odd_numbers=[]
for num in numbers:
  # Corrected even/odd check
  if num % 2 == 0:
    even_numbers.append(num)
  else:
    odd_numbers.append(num)
#Calculating average
average = sum(numbers)/len(numbers)
#round off the average
rounded_avg = round(average) # Assigned the rounded average
count_numbers = Counter(numbers)
#save result using os
# Corrected directory name to be consistent
if not os.path.exists("easy_reports"):
  os.mkdir("easy_reports")
#file path
file_path=os.path.join("easy_reports","number_report.txt")
#write data to a file
# Using 'with' statement for proper file handling and f-strings for formatting
with open(file_path,"w") as file:
  file.write(f"Generated Numbers:{numbers}\n")
  file.write(f"Even Numbers:{even_numbers}\n") # Corrected variable to even_numbers and used f-string
  file.write(f"Odd Numbers:{odd_numbers}\n")
  file.write(f"Average:{average}\n")
  file.write(f"Rounded Average:{rounded_avg}\n")
  file.write(f"Number Counts:{count_numbers}\n")
print(f"Report saved to {file_path}")
```
