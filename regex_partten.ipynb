```python
import re
text ="cat cot cut c9t c@t"
result=re.findall("c.t",text)
print(result)
```

```python
text="Hello world"
print(bool(re.search("^Hello",text))) # ^ to check starting
print(bool(re.search("world$",text))) # $ to check ending
```

```python
text="Hello world"
print(bool(re.search("Hello$",text)))
print(bool(re.search("^world",text)))
```

```python
text="helloooo"
result=re.findall("lo*",text)   # 0 or more
print(result)
```

```python
text="helloooo"
result=re.findall("lo+",text)  # one or more
print(result)
```

```python
text="color colour"
result=re.findall("colou?r",text)  # zero or one
print(result)
```

```python
text="apple ball cat"
result=re.findall("[abc]",text) # character set
print(result)
```

```python
text="my age is 30"
result=re.findall("[0-9]",text) # digit
print(result)
```

```python
text="my age is 30"
result=re.findall("[a-z]",text) # small alphabet
print(result)
```

```python
text="my age is 30"
result=re.findall("[A-Z]",text) # capital alphabet
print(result)
```

```python
text="my age is 30"
result=re.findall("[A-za-z]",text) # both small and capital
print(result)
```

```python
text="marks: 90"
result=re.findall(r"\d",text) # digit
print(result)
```

```python
text="marks: 90"
result=re.findall(r"\D",text) #  except digit
print(result)
```

```python
text="marks: 90"
result=re.findall(r"\w",text) # character
print(result)
```

```python
text="marks: 90"
result=re.findall(r"\W",text) # non character
```

```python
text="marks: 90"
result=re.findall(r"\s",text) # space
print(result)
```

```python
text="marks: 90"
result=re.findall(r"\S",text) # no space
print(result)
```

```python
text="phone : 75973423943756"
result=re.findall(r"\d{10}",text) #count
print(result)
```

```python
text=" i have a cat and a dog"
result=re.findall(r"cat|dog",text) #OR operator
print(result)
```

```python
text="aababa aba"
result=re.findall(r"(ab)+",text) #grouping
print(result)
```
