```python
#ATM SIMULATOR
file_name=("atm_data.txt")
balance=1000 # balance is the global variable that stores money
pin="1234"# pin is the global variable that stores atm pin
```

```python
#load data
def load_data():
  global balance,pin #global keyworx allowa us to modify outside variables
  try:
    file=open("atm_data.txt","r")
    lines=file.readlines()# reads all the lines from the file
    file.close()
    pin=lines[0].strip() # strip removes extra spaces from start and end
    balance=int(lines[1].strip())
  except:
    # if file does not exist or error occurs
    pass # will not do anything
```

```python
#CHECK BALANCE
def check_balance():
  print("your balance is:",balance)
```

```python
#SAVE DATA
def save_data():
  #opening the file in read mode
  file=open(file_name,"w")
  #write a pin into the file and go to next line
  file.write(pin+"\n")
  #write balance in string
  file.write(str(balance))
  #close the file
  file.close()
```

```python
#DEPOSIT MONEY
def deposit_money():
  # global allows changing original balance
  global balance
  try:
    amount=int(input("entre the amount to deposit:"))
    balance= balance+amount
    #save updated balance to the file
    save_data()
    print("money desposited sucessfully")
  except:
    print("please enter numbers only")
```

```python
#WITHDRAW MONEY
def withdraw_money():
  global balance
  try:
    amount=int(input("entre the amount to withdraw:"))
    if amount>balance:
      print("insufficient balance")
    else:
      balance= balance-amount
      #save updated balance to the file
      save_data()
      print("please collect ur cash")
  except:
    print("please enter numbers only")
```

```python
#CHANGE PIN
def change_pin():
  global pin
  #ask user for old pin
  old_pin = input("Enter your old pin: ")
  #check if old pin matches
  if old_pin == pin:
    #ask user for new pin
    new_pin = input("Enter your new pin: ")
    #update pin
    pin = new_pin
    #save updated pin to the file
    save_data()
    print("Pin changed successfully")
  else:
    print("Incorrect pin")
```

```python
#MAIN PROGRAM
def main():
  #load data from the file
  load_data()
  #ask user to enter a pin
  user_pin=input("enter the pin: ")
  if user_pin!=pin:
    print("incorrect pin")
    return #exit the function
  else:
    while True:
      print("----------------ATM MENU---------------")
      print("1. check balance")
      print("2. deposit money")
      print("3. withdraw money")
      print("4. change pin")
      print("5. exit")

      choice=input("enter your choice: ")
      if choice=="1":
        check_balance()
      elif choice=="2":
        deposit_money()
      elif choice=="3":
        withdraw_money()
      elif choice=="4":
        change_pin()
      elif choice=="5":
        print("thank you for using ATM")
        break
      else:
        print("invalid choice")
main()

```
