```python
#LOAD_HIGH_SCORE
def load_high_score():
  global high_score
  try:
    file=open("high_score.txt","r")
    high_score=int(file.read())
    file.close()
  except:
    # if file does not exist or error occurs
    pass

#SAVE_HIGH_SCORE
def save_high_score():
  file=open("high_score.txt","w")
  file.write(str(high_score))
  file.close()

#PLAY QUIZ
def play_quiz():
  global score
  question=["prime minister of india","national bird of india","pink city of india"]
  answers=["narandra modi","peacock","jaipur"]
  for i in range(len(question)):
    print("\nQ",question[i])
    user_answer=input("enter answer: ")
    if user_answer==answers[i]:
      score=score+1
      print("correct")
    else:
      print("wrong")

#MAIN_PROGRAM
def main():
  global high_score
  global score # Declare score as global
  score = 0     # Initialize score to 0 for each game
  high_score = 0 # Initialize high_score to 0
  load_high_score()
  name=input("enter name: ")
  play_quiz()
  print("\n quiz over")
  print(name,"your score is:",score)
  if score>high_score:
    high_score=score
    save_high_score()
  else:
    print("high score is:",high_score)
main()
```
